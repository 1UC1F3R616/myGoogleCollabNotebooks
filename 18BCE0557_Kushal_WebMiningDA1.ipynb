```
import requests
from bs4 import BeautifulSoup as soup

url = "https://en.wikipedia.org/wiki/Web_mining"
response = requests.get(url)
page_soup = soup(response.text, 'html.parser')

## Now cleaning tags
#page_text = page_soup.get_text()
#print(page_text)
page_text = '' ## Ignoring the noisy headings and other tags as later I found that we are working with sentences
paras = page_soup.findAll('p')
for para in paras:
    page_text = page_text + para.text
print(page_text)
```

```
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))

normalized_text = page_text.lower()
tokenized_words = nltk.word_tokenize(normalized_text)
terms = [word for word in tokenized_words if word not in stop_words]

terms_frequency = nltk.FreqDist(terms)
print("Number of Terms: {}".format(len(terms_frequency)))
print(dict(terms_frequency))
```

```
from string import punctuation
stop_words = stop_words.union(set(punctuation)).union({' ', '``', "''", '.', ','})
terms = [term for term in terms if term not in stop_words]
terms_frequency = nltk.FreqDist(terms)
print("Number of Terms: {}".format(len(terms_frequency)))
print(dict(terms_frequency))
```

```
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
import pandas as pd

lemmatizer = WordNetLemmatizer()
stemmer = nltk.stem.SnowballStemmer('english')
terms_lemmatized = [lemmatizer.lemmatize(term) for term in terms]
terms_stemmed = [stemmer.stem(term) for term in terms]

import pandas as pd
df = pd.DataFrame(data = {'Original Term': terms, 'Stemmed Term': terms_stemmed, 'Lemmatized Term': terms_lemmatized}).set_index('Original Term')
df.head(10)
```

```
tokenized_words_lemmatized = [lemmatizer.lemmatize(term) for term in tokenized_words]
tokenized_words_stemmed = [stemmer.stem(term) for term in tokenized_words]
print('Length of Lemmatized Words: {}'.format(len(tokenized_words_lemmatized)))
print('Length of Stemmed Words: {}'.format(len(tokenized_words_stemmed)))
```

```
nltk.download('averaged_perceptron_tagger')
stop_words2 = stop_words.difference(set(punctuation).union({' ', '``', "''", '.', ','}))
sentences = nltk.sent_tokenize(page_text)
sentences_stopwords = []
sentences_pos_tags = []
for sentence in sentences:
  temp = [word for word in nltk.word_tokenize(sentence) if word in stop_words2]
  sentences_stopwords.append(temp)
  sentences_pos_tags.append([pos_tag[1] for pos_tag in nltk.pos_tag(temp)])

df = pd.DataFrame(data = {'Original Sentence':sentences, 'List of Stopwords':sentences_stopwords, 'POS-Tags':sentences_pos_tags}).set_index('Original Sentence')
df.head(10)
```

```
url1 = "https://en.wikipedia.org/wiki/Web_mining"
url2 = "https://en.wikipedia.org/wiki/Data_mining"

response1 = requests.get(url1)
page_soup1 = soup(response1.text, 'html.parser')

response2 = requests.get(url2)
page_soup2 = soup(response2.text, 'html.parser')


page_text1 = ''
paras = page_soup1.findAll('p')
for para in paras:
    page_text1 = page_text1 + para.text
print(page_text1)


page_text2 = ''
paras = page_soup2.findAll('p')
for para in paras:
    page_text2 = page_text2 + para.text
print(page_text2)
```

```
normalized_text1 = page_text1.lower()
tokenized_words1 = nltk.word_tokenize(normalized_text1)
terms1 = [word for word in tokenized_words1 if word not in stop_words]
s = ' '.join(terms1)
f = open("wm_d1.doc", "w+")
f.write(s)
f.close()

normalized_text2 = page_text2.lower()
tokenized_words2 = nltk.word_tokenize(normalized_text2)
terms2 = [word for word in tokenized_words2 if word not in stop_words]
s = ' '.join(terms2)
f = open("wm_d2.doc", "w+")
f.write(s)
f.close()
```

```
bool1 = []
bag1 = []
complete1 = []

bool2 = []
bag2 = []
complete2 = []
for term in terms1+terms2:
  if term in terms1:
    bool1.append(1)
    bag1.append(terms1.count(term))
  else:
    bool1.append(0)
    bag1.append(0)

  if term in terms2:
    bool2.append(1)
    bag2.append(terms2.count(term))
  else:
    bool2.append(0)
    bag2.append(0)

  complete1.append([ind for ind, term_ in enumerate(terms1) if term_ == term])
  complete2.append([ind for ind, term_ in enumerate(terms2) if term_ == term])

#print(complete1)
#print(complete2)
```

```
# Term-Document  incidence  matrix using  Boolean
df = pd.DataFrame(data = {'Term': terms1+terms2, 'Web Mining Doc':bool1, 'Data Mining Doc':bool2}).set_index('Term')
df.tail(20)
```

```
# Term-Document incidence matrix using Bag of Words
df = pd.DataFrame(data = {'Term': terms1+terms2, 'Web Mining Doc':bag1, 'Data Mining Doc':bag2}).set_index('Term')
df.tail(20)
```

```
# Term-Document incidence matrix using Complete representation
df = pd.DataFrame(data = {'Term': terms1+terms2, 'Web Mining Doc':complete1, 'Data Mining Doc':complete2}).set_index('Term')
df.tail(20)
```

```
## Writea python program  to  show  the  implementation  of  sentence  paraphrasing through  synonyms(retaining  semantic  meaning)for  the  following foursentences.Display  at  least  three  other  paraphrased  sentences  for  each  sentence  mentioned below.

sentence1 = "The quick brown fox jumps over the lazy dog".split()
sentence2 = "Obama and Putin met the previous week".split()
sentence3 = "At least 12 people were killed in the battle last week".split()
sentence4 = "I will go home and come back tomorrow".split()

from nltk.corpus import wordnet

def syno(word):
  synonyms = wordnet.synsets(word)
  total = []
  for ss in synonyms:
    total += ss.lemma_names()
  return total

def maker(sentence):
  new1 = []
  new2 = []
  new3 = []
  for word in sentence:
    offers = syno(word)
    #print(offers)
    if len(offers) == 0:
      new1.append(word)
      new2.append(word)
      new3.append(word)
    else:
      last_index = abs(len(offers)-1)
      new1.append(offers[1%last_index])
      new2.append(offers[2%last_index])
      new3.append(offers[3%last_index])
  
  print(' '.join(new1))
  print(' '.join(new2))
  print(' '.join(new3))

maker(sentence1)
print()
maker(sentence2)
print()
maker(sentence3)
print()
maker(sentence4)
```
